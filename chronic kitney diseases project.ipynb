```python
cat_col = [i for i in data.columns if data[i].dtype == 'O']
num_col = [i for i in data.columns if i not in cat_col]
```

```python
def mean_value_imp(data,column):
    mean_value = data[column].mean()
    data[column].fillna(value = mean_value,inplace=True)
```

```python
def mode_value_imp(data,column):
    mode_value = data[column].mode()[0]
    data[column] = data[column].fillna(mode_value)
```

from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

```python
models = []
models.append(("logistic Regression",LogisticRegression()))
models.append(("RandomForestClassifier", RandomForestClassifier()))
models.append(("KNeighborsClassifier",KNeighborsClassifier(n_neighbors=8)))
models.append(("SVM",SVC(kernel='linear')))
```

```python
for name,model in models:
    print(name,model)
    print()
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print(confusion_matrix(y_test,y_pred))
    print("\n")
    print("accurucy:",accuracy_score(y_test,y_pred))
    print("\n")
    print("presion score:",precision_score(y_test,y_pred))
    print("\n")
    print("recall score:",recall_score(y_test,y_pred))
    print("\n")
    print("f1score:",f1_score(y_test,y_pred))
```
